```python
import json
with open('./data.json', 'r') as f:
    data = json.load(f)
```

```python
from datasets import load_dataset

ds = load_dataset("TheFusion21/PokemonCards")
```

```python
from PIL import Image
import requests
from io import BytesIO

#Get generated images, ground truth images, and prompts for test dataset
generated_imgs = []
ground_truth_imgs = []
prompts = []
for img in os.listdir('./generated_imgs/')[0:100]:
  #Get generated image
  generated_imgs.append(Image.open(f'./generated_imgs/{img}'))
  #Get image sample
  img_number = img.split('.')[0][13:]
  img_id = data[img_number]['id']
  img_sample = ds['train'].filter(lambda x: x['id'] == img_id)
  #Store caption and ground truth image
  prompts.append(img_sample['caption'][0])
  response = requests.get(img_sample['image_url'][0])
  #
  img = Image.open(BytesIO(response.content)).resize((
      generated_imgs[0].width, generated_imgs[0].height)).convert('RGB')
  ground_truth_imgs.append(img)
```

```python
os.chdir('./../../../../fine-tuning-stable-diffusion-3/metrics')
```

```python
import metrics

test = metrics.Metrics(ground_truth_imgs, generated_imgs, prompts)
test.compute_all_metrics(output_csv = 'metrics_output.csv')
```
